#### Simple Linear Regression
- **Structure**: Includes one dependent variable (\(Y\)) and one independent variable (\(X\)).
- **Equation**: \( Y = \beta_0 + \beta_1 X + \epsilon \), where \(\beta_0\) is the intercept, \(\beta_1\) is the slope.
- **Behavior**: Captures a straight-line relationship between \(Y\) and \(X\).

#### Multiple Linear Regression
- **Structure**: Includes one dependent variable (\(Y\)) and multiple independent variables (\(X_1, X_2, \dots\)).
- **Equation**: \( Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_p X_p + \epsilon \).
- **Behavior**: Models more complex relationships with multiple predictors.

---

### Continuous vs. Indicator Variables

- **Continuous Variables**: Numeric variables directly scaled by their coefficients (e.g., height, income).
- **Indicator Variables**: Represent binary data (e.g., yes/no), encoded as \(0\) or \(1\). Their coefficients show the mean difference between categories.

---

### Interaction Terms
- **Definition**: Used when the effect of one variable depends on another. For variables \(X_1\) and \(X_2\), the interaction term is \(X_1 \times X_2\).
- **Equation**: \( Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 (X_1 \times X_2) + \epsilon \).
- **Impact**:
  - \(\beta_1\): Effect of \(X_1\) when \(X_2 = 0\).
  - \(\beta_2\): Effect of \(X_2\) when \(X_1 = 0\).
  - \(\beta_3\): Change in \(X_1\)'s effect for a one-unit increase in \(X_2\).

---

### Handling Non-Binary Categorical Variables

- **Encoding**:
  - **One-Hot Encoding**: Creates \(k-1\) dummy variables for \(k\) categories.
  - **Effect Coding**: Similar to one-hot but uses \(-1\) for the reference category.
- **Representation**: Encoded variables contribute coefficients that measure differences relative to the baseline category.

### Logistic Regression for Binary Outcomes

#### Use Case
Logistic regression is used for **binary outcomes** (e.g., success/failure) and models the log-odds of the probability of an event (\(P(Y=1)\)).

#### Model Structure
- **Equation**: \(\text{logit}(P) = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_p X_p\), where:
  \[
  P(Y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_p X_p)}}.
  \]
- Predictors:
  - **Continuous** (\(X_1\)): Included directly.
  - **Binary** (\(X_2\)): Encoded as \(0\) or \(1\).

---

### Interpreting Coefficients

- \(\beta_j\): Represents the change in **log-odds** of the outcome for a one-unit increase in \(X_j\).
- **Odds Ratio**: \(e^{\beta_j}\):
  - \(e^{\beta_j} > 1\): Positive association.
  - \(e^{\beta_j} < 1\): Negative association.

---

### Visualization: Additive vs. Interaction Models

#### Additive Model
- **Equation**: \(\text{logit}(P) = \beta_0 + \beta_1 X_1 + \beta_2 X_2\).
- **Behavior**: Effects of \(X_1\) and \(X_2\) are independent.
- **Plot**: Parallel curves when plotting probabilities against \(X_1\) for fixed \(X_2\).

#### Interaction Model
- **Equation**: \(\text{logit}(P) = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 (X_1 \times X_2)\).
- **Behavior**: The effect of \(X_1\) depends on \(X_2\).
- **Plot**: Diverging/converging curves based on the interaction term (\(\beta_3\)).

#### Steps to Plot
1. Compute predicted probabilities using:
   \[
   P = \frac{1}{1 + e^{-(\beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 (X_1 \times X_2))}}.
   \]
2. Plot \(P\) vs. \(X_1\) for fixed \(X_2\) values.

### Reconciling Low \(R^2\) and Significant Coefficients

   - Most variability in \(Y\) is unexplained (high noise), but predictors still show consistent effects on \(Y\).  
   - Low \(R^2\) reflects unexplained variance, while low p-values highlight consistent predictor impact.

2. **Large Sample Size**:  
   - Larger datasets reduce standard errors, making even small effects statistically significant.  
   - Small \(R^2\) and significant p-values coexist due to sample size.

3. **Weak Predictors**:  
   - Predictors have small but real effects.  
   - Collectively, they explain little variance (\(R^2\)), but individual effects are statistically significant.

4. **High Outcome Variability**:  
   - Unmeasured factors dominate variability in \(Y\).  
   - Predictors capture only a small portion, resulting in significant coefficients but low \(R^2\).

5. **Nonlinear Relationships**:  
   - Linear models may miss complex relationships, leading to low \(R^2\).  
   - Significant coefficients reflect captured linear trends.

6. **\(R^2\) Limitations**:  
   - For logistic regression or categorical outcomes, traditional \(R^2\) is less meaningful.  
   - Significant p-values indicate predictor relevance, even if \(R^2\) is low.

**Conclusion**: Significant coefficients suggest meaningful relationships, while low \(R^2\) reflects unexplained variability or model limitations.

- **Steps**: Predictions are made on the test set, and out-of-sample \( R^2 \) is computed.

- **Purpose**: Determine if added complexity improves out-of-sample predictions.
- **Key Point**: If test \( R^2 \) decreases, the model is likely overfitting.

### 1. Centering and Scaling Predictors

- **Centering**: Subtracts the mean of each predictor, reducing correlation between main effects and interaction terms.  
- **Scaling**: Divides predictors by their standard deviation, ensuring all predictors have similar ranges and stabilizing coefficient estimates.  
- **Effect**: Reduces multicollinearity by decreasing redundancy and improving numerical stability in models with interaction or polynomial terms.

---

### 2. Multicollinearity with Interaction Terms

- **Persistence**: Interaction terms (e.g., \(X_1 \times X_2\)) remain correlated with their base predictors, even after centering.  
- **Higher-Order Terms**: Polynomial terms also introduce dependencies that are hard to eliminate.  
- **Impact**: Multicollinearity from interaction terms leads to unstable coefficients and limits model interpretability.

---

### 3. Impact on Generalization

- **Multicollinearity Issues**: Causes unstable predictions, overfitting, and poor out-of-sample performance.  
- **After Centering/Scaling**: Stabilizes coefficients, reducing overfitting risks, but residual multicollinearity can still hinder generalization.

---

### 4. Condition Number

- **Definition**: Ratio of the largest to smallest singular value of the design matrix; high values signal multicollinearity.  
- **Low Condition Number**: Indicates stable, well-conditioned predictors.  
- **High Condition Number**: Suggests remaining multicollinearity, leading to unreliable coefficient estimates.

---

### Key Takeaways

1. Centering/scaling helps reduce multicollinearity but cannot fully resolve it in complex models.  
2. Interaction terms inherently introduce correlations that destabilize predictions.  
3. High condition numbers highlight lingering multicollinearity, requiring regularization or alternative modeling techniques for better generalization.

````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# Generate synthetic data for demonstration
np.random.seed(42)  # Fix seed for data reproducibility, not for train-test split
X = np.random.rand(100, 2)  # 100 samples, 2 features
y = 3 * X[:, 0] + 2 * X[:, 1] + np.random.randn(100) * 0.1  # Linear relationship with noise

# Initialize lists to store metrics
train_r2_scores = []
test_r2_scores = []

# Number of iterations
n_iterations = 100

# Loop for repeated train-test splits and model fitting
for _ in range(n_iterations):
    # Split data (no fixed random state for variability)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    
    # Fit the model
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    # Evaluate metrics
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    train_r2_scores.append(r2_score(y_train, y_train_pred))  # In-sample performance
    test_r2_scores.append(r2_score(y_test, y_test_pred))    # Out-of-sample performance

# Convert to DataFrame for analysis
metrics_df = pd.DataFrame({
    "Train R^2": train_r2_scores,
    "Test R^2": test_r2_scores
})

# Visualize the distribution of metrics
plt.figure(figsize=(10, 6))
plt.hist(metrics_df["Train R^2"], bins=20, alpha=0.5, label='Train R^2')
plt.hist(metrics_df["Test R^2"], bins=20, alpha=0.5, label='Test R^2')
plt.axvline(np.mean(train_r2_scores), color='blue', linestyle='--', label='Mean Train R^2')
plt.axvline(np.mean(test_r2_scores), color='orange', linestyle='--', label='Mean Test R^2')
plt.xlabel('R^2 Score')
plt.ylabel('Frequency')
plt.title('Distribution of R^2 Scores Across Iterations')
plt.legend()
plt.show()
````

### Analyzing Predictive Performance Across Subsets

#### Purpose
Training a model on one subset (e.g., Generation 1) and testing on another (e.g., Generation 6) evaluates its ability to **generalize** to unseen data with different characteristics.

---

### Key Metrics
1. **In-Sample \( R^2 \):**
   - Measures how well the model fits the training data (e.g., Generation 1).
   - High \( R^2 \) indicates a good fit for the training subset.

2. **Out-of-Sample \( R^2 \):**
   - Assesses the model’s performance on unseen data (e.g., Generation 6).
   - A low \( R^2 \) here suggests poor generalization due to overfitting or differences in the data.

---

### Interpretation of Results
1. **High In-Sample \( R^2 \), Low Out-of-Sample \( R^2 \):**
   - **Explanation**: Overfitting to training data or significant differences between subsets.
   - **Solution**: Simplify the model or use regularization.

2. **Similar In-Sample and Out-of-Sample \( R^2 \):**
   - **Explanation**: The model generalizes well across subsets.
   - **Solution**: No major changes needed; model is robust.

3. **Moderate Out-of-Sample \( R^2 \):**
   - **Explanation**: Subset differences exist but do not drastically impact generalization.
   - **Solution**: Investigate and align feature distributions.

---

### Insights and Mitigation
- **Identify Data Distribution Differences:** Analyze feature distributions between subsets.
- **Regularization:** Penalize complexity to improve generalization.
- **Cross-Validation:** Validate on multiple subsets to ensure robustness.
